# التقرير الأولي


### أوجد إجمالي إنتاج الحليب لعام 2023

```python
%%sql
select sum(new_values) as total_Milk_Production
from milkProdu
where Year = 2023 ;
```

 ### أوجد إجمالي إنتاج القهوة لعام 2015

```python
%%sql
select sum(new_values) as Total_Coff_Prod2015
from coffeeProdu
where Year = 2015;
```

 ### أوجد متوسط إنتاج العسل لعام 2022

```python
%%sql
select avg(new_values) as Aver_Hny_Produ2022
from honeyProdu
where Year = 2022;
```

 ### أوجد أعلى إنتاج للزبادي لعام 2022

```python
%%sql
select max(new_values) as max_youg_Prod2022
from YougertProdu
where Year = 2022;
```

 ### أوجد الولايات التي تنتج الحليب والعسل  
  ### لعام 2022

```python
%%sql
select distinct milkProdu.State_ANSI
from milkProdu 
inner join honeyProdu on milkProdu.State_ANSI = honeyProdu.State_ANSI
where milkProdu.Year = 2022 and honeyProdu.Year = 2022;
     
```

```python
%%sql
select sum(YougertProdu.new_values)
from YougertProdu
where YougertProdu.Year = 2022 and State_ANSI in (select distinct State_ANSI from cheeseProdu where cheeseProdu.Year= 2022);
```
